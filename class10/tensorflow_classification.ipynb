```python
import pandas as pd
data = pd.read_csv("./drive/MyDrive/Colab-Notebooks/ml-363/advertising.csv")
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x = data[["Daily Time Spent on Site", "Age", "Area Income", "Daily Internet Usage", "Male"]]
y = data["Clicked on Ad"]

scaler = StandardScaler()
sc_x = scaler.fit_transform(x)

train_x, test_x, train_y, test_y = train_test_split(sc_x, y, test_size=0.2, random_state=10)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
print(sc_x.shape)
```

```python
model = keras.Sequential([
  keras.layers.Dense(2,activation="relu",input_shape=(5,)),
  keras.layers.Dense(3,activation="relu"),
  keras.layers.Dense(1,activation='sigmoid')      
])
```

```python
bse = keras.losses.BinaryCrossentropy()
evaluation = keras.metrics.BinaryAccuracy()
optim = keras.optimizers.Adam()
```

```python
model.compile(loss=bse, optimizer=optim, metrics= evaluation)
```

```python
batch_size = 32
epochs = 500
model.fit(train_x, train_y, batch_size=batch_size, shuffle=True ,epochs=epochs)
```

```python
model.evaluate(test_x, test_y, batch_size = batch_size)
```
