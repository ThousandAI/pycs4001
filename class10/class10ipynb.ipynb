# **Word Embedding**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = ["It's raining outside.",
          "Studying at school is far more efficient than at home.",
          "I clean school seats every day"]
```

```python
vectorizer = TfidfVectorizer()
wordvec = vectorizer.fit_transform(corpus)
```

```python
vectorizer.get_feature_names_out()
```

```python
print(wordvec.toarray())
```
